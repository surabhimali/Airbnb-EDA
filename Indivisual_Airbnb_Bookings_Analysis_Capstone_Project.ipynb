We will also list out the average price for each type of room per neighbourhood so that tourists can plan based on the budget.


```python
df.head()
plt.figure(figsize=(12,8))
df = df[df['minimum_nights']==1]
df3 = df.groupby(['room_type','neighbourhood_group'])['price'].mean().sort_values(ascending=True)
df3.plot(kind='bar')
plt.title('Average Price for rooms in neighbourhood group')
plt.ylabel('Average Daily Price')
plt.xlabel('Neighbourhood Group')
plt.show()
```

```python
print('List of Average Price per night based on the neighbourhood group')
pd.DataFrame(df).sort_values(by='room_type')
```

***Which neighbourhood has the highest and lowest price?***

---



Now we will try to understand which locality has the highest price margin in terms of night being spend by traveller on the basis of Neighbourhood and number of reviews.
